```python
# These are the standard imports for CS 111. 
# This list may change as the quarter goes on.

import os
import math
import time
import struct
import numpy as np
import numpy.linalg as npla
import scipy
import scipy.sparse.linalg as spla
from scipy import sparse
from scipy import linalg
from scipy import integrate
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import axes3d
%matplotlib tk
```

```python
# demo of ode23tx.m = rk23.py
# big feeling that this will be on the test because he is taking so much time on it, minus the plotting
def f(t, y):
    """function to be integrated to solve an ODE or a system of ODEs
    Input:
      t is a scalar time
      y is a vector of variables
    Output:
      ydot is the vector dy/dt
    """
    ydot = (2/math.sqrt(math.pi))*math.exp(-t**2)
    return ydot

tspan = (0,2)
yinit = [0]

sol = integrate.solve_ivp(fun = f, t_span = tspan, y0 = yinit, method = 'RK23')
#more intial condtions for higher order
#RK23 some french method to solve using 3rd order approximation like RK 45, 
#or ODE 1 and ODE2 those last two is the ones we made ourselfes

%matplotlib inline
plt.plot(sol.t, sol.y[0], '.', label='ode solution')
tt = np.linspace(0, 2, 100)
plt.plot(tt, np.erf(tt), label='erf(t)')
plt.legend()
plt.xlabel('t')
plt.ylabel('y')
plt.title('erf(x) ODE')
```

```python
# demo of ode23tx.m = rk23.py
# big feeling that this will be on the test because he is taking so much time on it, minus the plotting
def f(t, y):
    """function to be integrated to solve an ODE or a system of ODEs
    Input:
      t is a scalar time
      y is a vector of variables
    Output:
      ydot is the vector dy/dt
    """
    ydot = -50*y
    return ydot

tspan = (0,2)
yinit = [-2]

sol = integrate.solve_ivp(fun = f, t_span = tspan, y0 = yinit, method = 'RK23')
#more intial condtions for higher order
#RK23 some french method to solve using 3rd order approximation like RK 45, 
#or ODE 1 and ODE2 those last two is the ones we made ourselfes

%matplotlib inline
plt.plot(sol.t, sol.y[0], 'o-', label='ode solution')
tt = np.linspace(0, 2, 1000)
plt.legend()
plt.xlabel('t')
plt.ylabel('y')
plt.title('erf(x) ODE')

plt.gca().set_ylim([-.00002,.00002])
```
